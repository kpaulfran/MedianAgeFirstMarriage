Preprocesamiento de Datos

```python
#Se llaman a todas las librerias necesarias

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/insurance.csv')
df.info()
```

```python
df.head()
```

```python
y = df['charges']
X = df.drop(columns = 'charges')
```

```python
y.head()
```

```python
X.head()
```

```python
X_train, X_test, y_train,  y_test = train_test_split(X,y, random_state = 42, test_size=0.2)
```

```python
X_train.info()
```

Datos:

- Age: Numero Entero
- Sex: Texto
- BMI: float
- Children: Entero
- Smoker: Texto
- Region: Texto

```python
X_train['sex'].unique()
```

```python
X_train['sex'].replace({'female': 0, 'male': 1}, inplace = True)
X_train['sex'].unique()
```

```python
X_test['sex'].replace({'female': 0, 'male': 1}, inplace = True)
```

```python
X_train['smoker'].unique()
```

```python
X_train['smoker'].replace({'yes': 1, 'no': 0}, inplace = True)
X_test['smoker'].replace({'yes': 1, 'no': 0}, inplace = True)
X_train['smoker'].unique()
```

```python
X_train['region'].unique()
```

```python
cat_selector = make_column_selector(dtype_include='object')
```

```python
cat_selector(X_train)
```

```python
train_cat_data = X_train[cat_selector(X_train)]
test_cat_data = X_test[cat_selector(X_test)]
train_cat_data
```

```python
ohe_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

ohe_encoder.fit(train_cat_data)

train_ohe = ohe_encoder.transform(train_cat_data)
test_ohe = ohe_encoder.transform(test_cat_data)
train_ohe
```

```python
#Convertir a DataFrame

ohe_column_names = ohe_encoder.get_feature_names_out(train_cat_data.columns)
train_ohe = pd.DataFrame(train_ohe, columns=ohe_column_names)
test_ohe = pd.DataFrame(test_ohe, columns=ohe_column_names)
train_ohe
```

```python
num_selector = make_column_selector(dtype_include='number')

train_nums = X_train[num_selector(X_train)].reset_index(drop=True)
test_nums = X_test[num_selector(X_test)].reset_index(drop=True)
```

```python
scaler = StandardScaler()
scaler.fit(train_nums)

train_scaled = scaler.transform(train_nums)

test_scaled = scaler.transform(test_nums)

train_scaled[:5]
```

```python
X_train_scaled = pd.DataFrame(train_scaled, columns=train_nums.columns)
X_train_scaled.head()
```

```python
X_test_scaled = pd.DataFrame(test_scaled, columns=test_nums.columns)
X_test_scaled.head()
```

```python
X_train_processed = pd.concat([X_train_scaled, train_ohe], axis=1)
X_test_processed = pd.concat([X_test_scaled, test_ohe], axis=1)
X_train_processed
```
